### Task - 1 : Restaurant Ratings

##### Analyze the distribution of aggregate ratings and determine the most common rating range.

```python
# Extract aggregate ratings
aggregate_ratings = df['Aggregate rating']

# Plot histogram
plt.hist(aggregate_ratings, bins=10, color='skyblue')

# Customize plot
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()

# Determine the most common rating range
most_common_range = aggregate_ratings.mode()[0]
print(f"The most common rating range is {most_common_range}.")
```

##### Calculate the average number of votes received by restaurants.

```python
# Calculate the average number of votes received by restaurants
average_votes = df['Votes'].mean()
print(f"Average number of votes received by restaurants: {average_votes:.2f}")
```

### Task-2 : Cuisine Combination

##### Identify the most common combinations of cuisines in the dataset.

```python

# Drop rows with missing values in the 'Cuisines' column
df.dropna(subset=['Cuisines'], inplace=True)

# Extract the 'Cuisines' column and split the comma-separated values
cuisine_combinations = df['Cuisines'].str.split(', ')

# Create an empty dictionary to store the count of each cuisine combination
cuisine_count = {}

# Iterate over each row in the 'Cuisines' column
for cuisines in cuisine_combinations:
    # Sort the list of cuisines to ensure consistency in counting
    cuisines.sort()
    # Convert the sorted list of cuisines back to a string to use it as a dictionary key
    key = ', '.join(cuisines)
    # Update the count of the cuisine combination in the dictionary
    cuisine_count[key] = cuisine_count.get(key, 0) + 1

# Sort the dictionary by the count of each cuisine combination in descending order
sorted_cuisine_count = sorted(cuisine_count.items(), key=lambda x: x[1], reverse=True)

# Print the most common combinations of cuisines
print("Most common combinations of cuisines:")
for cuisine, count in sorted_cuisine_count:
    print(f"{cuisine}: {count}")
```

##### Determine if certain cuisine combinations tend to have higher ratings.

```python

# Drop rows with missing values in the 'Cuisines' or 'Aggregate rating' columns
df.dropna(subset=['Cuisines', 'Aggregate rating'], inplace=True)

# Extract the 'Cuisines' and 'Aggregate rating' columns
cuisine_ratings = df[['Cuisines', 'Aggregate rating']]

# Create an empty dictionary to store the total rating and count of each cuisine combination
cuisine_rating_count = {}

# Iterate over each row in the dataframe
for index, row in cuisine_ratings.iterrows():
    cuisines = row['Cuisines'].split(', ')
    rating = row['Aggregate rating']
    # Sort the list of cuisines to ensure consistency in counting
    cuisines.sort()
    key = ', '.join(cuisines)
    if key in cuisine_rating_count:
        cuisine_rating_count[key][0] += rating
        cuisine_rating_count[key][1] += 1
    else:
        cuisine_rating_count[key] = [rating, 1]

# Calculate the average rating for each cuisine combination
average_ratings = {key: value[0] / value[1] for key, value in cuisine_rating_count.items()}

# Sort the dictionary by the average rating of each cuisine combination in descending order
sorted_average_ratings = sorted(average_ratings.items(), key=lambda x: x[1], reverse=True)

# Print the average rating for each cuisine combination
print("Average rating for each cuisine combination:")
for cuisine, rating in sorted_average_ratings:
    print(f"{cuisine}: {rating:.2f}")
```

### Task - 3 : Geographic Analysis

##### Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
# Plot the locations on a map
plt.figure(figsize=(10, 8))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Locations of Restaurants')
plt.grid(True)
plt.show()
```

##### Identify any patterns or clusters of restaurants in specific areas.

```python
from sklearn.cluster import KMeans

# Extract longitude and latitude coordinates
X = df[['Longitude', 'Latitude']]

# Perform K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # You can adjust the number of clusters as needed
df['Cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(10, 8))
for cluster in df['Cluster'].unique():
    cluster_df = df[df['Cluster'] == cluster]
    plt.scatter(cluster_df['Longitude'], cluster_df['Latitude'], label=f'Cluster {cluster}', alpha=0.5)

# Plot the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', label='Cluster Centers')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clusters of Restaurants')
plt.legend()
plt.grid(True)
plt.show()
```

### Task - 4 : Restaurant Chains

##### Identify if there are any restaurant chains present in the dataset.

```python
# Convert restaurant names to lowercase for consistency
df['Restaurant Name'] = df['Restaurant Name'].str.lower()

# Group restaurants by name and count the number of occurrences
restaurant_counts = df['Restaurant Name'].value_counts()

# Filter restaurants with multiple locations (chains)
restaurant_chains = restaurant_counts[restaurant_counts > 1]

# Print the identified restaurant chains
print("Restaurant Chains:")
for chain, count in restaurant_chains.items():
    print(f"{chain}: {count} locations")
```

##### Analyze the ratings and popularity of different restaurant chains.

```python
# Convert restaurant names to lowercase for consistency
df['Restaurant Name'] = df['Restaurant Name'].str.lower()

chain_stats = df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Restaurant ID': 'count'  
})

# Rename the column to represent the number of locations
chain_stats.rename(columns={'Restaurant ID': 'Number of Locations'}, inplace=True)

# Sort the chains by average rating and total votes in descending order
chain_stats = chain_stats.sort_values(by=['Aggregate rating', 'Votes'], ascending=False)

print("Top Restaurant Chains:")
print(chain_stats.head(10))
```
